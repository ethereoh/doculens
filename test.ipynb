```python
import random
import pandas as pd

# ----- PROMPTS ENGINEER -----
def chain_of_thought(df:pd.DataFrame, n_samples:int=30) -> list[str]: 
    rand_idx = random.sample(range(0, len(df)), n_samples)
    result_df = df.iloc[rand_idx][['question', 'context']]
    questions = result_df.question.to_list()
    contexts = result_df.context.to_list()
    results = []

    for question, context in zip(questions, contexts): 
        result = question + ": " + context[2: -2] # Remove open/close brackets and quote ([] and '')
        results.append(result)
    return results
    

df = pd.read_csv('./db/train.csv', index_col=False)
results = chain_of_thought(df)
for res in results: 
    print(res)
    print("====")
```
